# Checking the sizes of the downloaded files

- 'files1.csv' is a list of the files actually downloaded
  - aws s3 ls s3://laelgelctweets/ --human-readable --summarize > files1.txt
  - aws s3 cp files1.txt s3://gelc/
  - Formatting 'file1.txt' into 'file1.csv' via VS Code by replacing regex \s+ by ','
- 'files2.csv' is a list of the files obtained from the original archive
  - Formatting 'file2.txt' into 'file2.csv' via VS Code by replacing regex \s+ by ','

## Importing the required packages

```python
import pandas as pd
```

## Processing 'file1.csv'

```python
df_files1 = pd.read_csv('files1.csv', header=None)

df_files1.drop(df_files1.columns[[0, 1]], axis=1, inplace=True)

df_files1 = df_files1[[4, 2, 3]]

df_files1.columns = pd.RangeIndex(df_files1.columns.size)

df_files1.sort_values(by=0, ascending=True, inplace=True)

df_files1.reset_index(drop=True, inplace=True)

df_files1
```

## Processing 'file2.csv'

```python
df_files2 = pd.read_csv('files2.csv', header=None)

df_files2.drop(df_files2.columns[[1, 2, 3]], axis=1, inplace=True)

df_files2.columns = pd.RangeIndex(df_files2.columns.size)

df_files2.sort_values(by=0, ascending=True, inplace=True)

df_files2.reset_index(drop=True, inplace=True)

df_files2
```

## Concatenating the information for comparison

```python
df_files3 = pd.concat([df_files1, df_files2], axis=1)

df_files3.columns = pd.RangeIndex(df_files3.columns.size)

df_files3 = df_files3[[0, 1, 2, 4, 3]]

df_files3.columns = ['filename-destination', 'filesize-destination', 'unit-destination', 'filesize-origin (GB)', 'filename-origin']

df_files3.to_csv('files3.csv')

df_files3
```
